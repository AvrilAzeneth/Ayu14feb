# Tercera práctica: Codificación y decodificación de una cadena:
El objetivo de la práctica, es: dado un alfabeto, se puede hacer una asignación de este a cadenas binarias, y viceversa, dado una cadena binaria, con el mismo
alfabeto, se busca hacere una asignación a cadenas de textos o cualquier tipo de dato incluido en un documento con extensión ".xlsx"


```python
# Primero importamos las bibliotecas.
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')


def leer_archivo(arch_excel):
  """Esta funcion se encarga, como su nombre indica, de leer un archivo de excel."""
  arch = pd.read_excel(arch_excel, dtype = {'Binary coding' : str})
  return arch

def extraer_columnas(datos):
  """Extraemos por nombre de columnas (es muy importante que se llamen literal como
  en el archivo)."""
  letras = list(datos["Letters"])
  cod_bin = list(datos["Binary coding"])
  dictionary = dict(zip(letras, cod_bin))

  return dictionary

class trie:
    def __init__(self, char, left=None, right=None):
        self.char = char
        self.left = left
        self.right = right
        self.father = None
        self.bit = None


def get_codes(code, node, bits):
    """Este metodo se encarga de recorrer el trie y va construyendo los codigos en
    binarios para cada caracter. Se hace un llamado de forma recursiva, hasta que
    se alcanza una hoja respectiva del trie."""
    if len(node.char) != 1:
        get_codes(code, node.left, bits + '1')
        get_codes(code, node.right, bits + '0')
    else:
      return


def encoding(text, codes):
    """Este metodo se encarga de codificar el texto recibido"""
    binary_text = ''.join(codes[char] for char in text)
    return binary_text


def invert_codes(codes):
    """En este metodo se invierte el diccionario de codigos,es decir: intercambia
    las claves por los valores."""
    inverted_codes = {v: k for k, v in codes.items()}
    return inverted_codes


def decoding(b_text, codes):
    """Aqui se hace la decodificacion de la cadena binaria respectiva. """
    inverted_codes = invert_codes(codes)
    decode_text = ''
    tmp = ''

    for bit in b_text:  # Este for itera sobre cada bit en la cadena binaria.
        tmp += bit
        if tmp in inverted_codes:
            decode_text += inverted_codes[tmp]
            tmp = ''

    return decode_text

def main():
    """Funcion principal, en el que se mandan llamar los principales metodos
    de codificacion y decodificacion con sus respectivas consecuencias."""
    file = '/content/drive/MyDrive/Progra2024-2/Code.xlsx'
    datos = leer_archivo(file)
    letras = extraer_columnas(datos)

    # La variable codes sera el diccionario que se obtiene del excel dado.
    codes = letras
    print('Alfabeto obtenido:', codes)

    print("Escribe el texto que deseas codificar")
    # La variable texto sera el mensaje codificado.
    texto = input (' ')

    print("Escribe el mensaje binario que deseas decifrar")
    # La siguiente variable sera el mensaje en binario.
    b_text = input (' ')
    binary_text = encoding(texto, codes)
    print('Texto codificado:', binary_text)

    decode_text = decoding(b_text, codes)
    print('Texto decodificado:', decode_text)

if __name__ == '__main__':
    main()
```
